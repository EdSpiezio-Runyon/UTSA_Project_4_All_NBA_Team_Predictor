```python
# Dependencies
import pandas as pd
```

# All NBA Data

```python
all_nba_df = pd.read_csv("Resources/End of Season Teams.csv")
all_nba_df
```

```python
# filter rows where year is below 2016
modern_all_nba = all_nba_df[all_nba_df['season'] >= 1980]
modern_all_nba
```

```python
# drop unnecessary columns
modern_all_nba = modern_all_nba.drop(['lg', 'position', 'birth_year', 'player_id', 'age', 'tm', 'player', 'season', 'number_tm'], axis=1)
modern_all_nba
```

```python
modern_all_nba = modern_all_nba[modern_all_nba['type'] == 'All-NBA']
modern_all_nba
```

# Player Total Data

```python
player_totals_df = pd.read_csv("Resources/Player Totals.csv")
player_totals_df
```

```python
modern_player_totals = player_totals_df[player_totals_df['season'] >= 2020]
modern_player_totals
```

```python
modern_player_totals = modern_player_totals.drop(['birth_year'], axis=1)
modern_player_totals
```

# Team Data

```python
team_summaries_df = pd.read_csv("Resources/Team Summaries.csv")
team_summaries_df
```

```python
modern_team_summaries = team_summaries_df[team_summaries_df['season'] >= 2020]
win_loss = modern_team_summaries[['season', 'abbreviation', 'w', 'l']]
win_loss['games_played'] = win_loss['w'] + win_loss['l']
win_loss['win_percent'] = win_loss['w'] / win_loss['games_played']
win_loss = win_loss.dropna()
win_percent = win_loss[['season', 'abbreviation', 'win_percent', 'games_played']]
win_percent.rename(columns={'abbreviation':'tm'}, inplace=True)
win_percent
```

# Merge Data

```python
merged_data = pd.merge(modern_all_nba, modern_player_totals, on=['seas_id'], how='right')
merged_data
```

```python
# create a new column that indicates whether or not the player was an All-NBA player
merged_data['all_nba'] = merged_data['type'].notnull().astype(int)
merged_data
```

```python
merged_data = merged_data.drop(['type'], axis=1)
merged_data
```

```python
all_merged = pd.merge(merged_data, win_percent, on=['season', 'tm'], how='left')
all_merged
```

# Visualization

```python
previous_years = all_merged[all_merged['season'] != 2023]
previous_years['ppg'] = previous_years['pts'] / previous_years['g']
previous_years['all_nba'] = previous_years['all_nba'].replace({0: False, 1: True})
previous_years.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
colors = ['red', 'blue']
plt.xlabel("Win Percentage")
plt.ylabel('Points Per Game')
plt.scatter(previous_years['win_percent'], previous_years['ppg'], c=previous_years['all_nba'].apply(lambda x: colors[x]), alpha=previous_years['all_nba'].apply(lambda x: 0.3 if x==0 else 1.0))
plt.show()
```

```python
import plotly.express as px

fig = px.scatter(previous_years, x = "win_percent", y = 'ppg', color = "all_nba", category_orders={"all_nba": [True, False]},
                 color_continuous_scale=['red', 'blue'],
                 labels = {"all_nba": "All-NBA"})
fig.update_layout(xaxis_title = "Win Percentage", yaxis_title = "Points Per Game", 
                  title= 'Scoring and Winning relation with All-Star Selection (1980-2022)', yaxis_range = [0,40])
fig.show()
```

```python
test_data = merged_data.drop(['player_id', 'lg', 'tm', 'pos', 'player'], axis=1)
test_data = test_data[test_data['season'] <= 2022]
test_data
```

```python
test_data = test_data.dropna()
test_data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
# Split the data into X_train, X_test, y_train, y_test
X = test_data.drop("all_nba", axis=1)
y = test_data["all_nba"]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Scale the features using a standard scaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train a Logistic Regression model and print the model score
classifier = LogisticRegression(max_iter=10000)

classifier.fit(X_train_scaled, y_train)

print(f"Training Data Score: {classifier.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test)}")
```

```python
# Make predictions on new data
prediction_data = merged_data[merged_data['season'] == 2023]
prediction_data = prediction_data.drop(['player_id', 'lg', 'tm', 'pos', 'player', 'all_nba'], axis=1)
prediction_data = prediction_data.dropna()
prediction_data_scaled= scaler.transform(prediction_data)
predictions = classifier.predict(prediction_data_scaled)
```

```python
# Use the trained logistic regression model to make predictions
proba = classifier.predict_proba(prediction_data_scaled)
```

```python
# Add the All-NBA probability to the prediction_data dataframe
prediction_data['All-NBA Probability'] = proba[:, 1]
```

```python
# Round the All-NBA Probability column to the nearest hundredth of a percent
prediction_data['All-NBA Probability'] = (prediction_data['All-NBA Probability'] * 100).round(2)

# Sort the prediction_data dataframe by All-NBA probability in descending order
prediction_data = prediction_data.sort_values(by='All-NBA Probability', ascending=False)
```

```python
predicted_players = prediction_data.loc[predictions == 1]
predicted_players
```

```python
# Train a Random Forest Classifier model and print the model score
clf = RandomForestClassifier(n_estimators=500).fit(X_train, y_train)

print(f'Training Score: {clf.score(X_train, y_train)}')
print(f'Testing Score: {clf.score(X_test, y_test)}')
```

```python
# Scale the new data using the same scaler used on the training data
prediction_data_scaled = scaler.transform(prediction_data)

# Use the trained Random Forest model to make predictions
predictions = classifier.predict(prediction_data_scaled)

# Print the predicted All-NBA players
predicted_players = prediction_data[predictions == 1]
predicted_players
```
